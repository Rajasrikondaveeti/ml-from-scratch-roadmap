Formatting with PlaceHolders


```python
#% = string formatting operator

print('Inject your name %s here' %'Shreyas')
```

```python
print(" I'll put some here : %s and  here: %s" %('if','any'))
```

```python
x = 'if'
y = 'any'

print(" I'll put some here : %s and  here: %s" %(x,y))
```

```python
print(" I'll put some here : %s and  here: %r" %('if','any'))
```

#Padding and Precision of Floating point Numbers

* %5.2f : 5 is min no. of characters the string should contain. .2f stands for how many numbers to show past the decimal point

```python
print('My floating point number is %5.3f' %(12.1567))
```

```python
print('My floating point number is %1.0f' %(12.1567))
```

```python
print('My floating point number is %30.1f' %(12.1567))
```

# .format()

```python
print('This is a string with an {}'.format('inserted string'))
```

```python
print('I have {}, {} and {}'.format('car','scooter','bike'))
```

```python
print('I have {2}, {1} and {0}'.format('car','scooter','bike'))
```

```python
print('I have {x}, {y} and {z}'.format(y = 'car',z= 'scooter',x= 'bike'))
```
