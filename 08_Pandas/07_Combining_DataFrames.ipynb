# Concatenation

```python
data_one = {
    'A' : ['A0', 'A1','A2','A3'],
    'B' : ['B0','B1','B2','B3']
}


data_two = {
    'C' : ['C0','C1','C2','C3'],
    'D' : ['D0','D1','D2','D3']
}
```

```python
one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
#concatenate along rows (axis 0)

pd.concat([one,two],axis=0)
```

```python
pd.concat([one,two],axis=1) #concatenate along columns (axis 1)
```

```python
data_three = {
    'A' : ['C0','C1','C2','C3'],
    'B' : ['D0','D1','D2','D3']
}

three = pd.DataFrame(data_three)
three
```

```python
pd.concat([one,three]) #default axis is 0
```

# Merge

```python
reg = pd.DataFrame(
    {
        'reg_id' : [1,2,3,4],
        'name' : ['A','B','C','D']
    }
)
```

```python
logins = pd.DataFrame(
    {
        'log_id' : [1,2,3,4],
        'name' : ['X','A','Y','B']
    }
)
```

```python
reg
```

```python
logins
```

```python
pd.merge(reg,logins,how='inner',on='name')
```

```python
pd.merge(reg,logins,how='outer',on='name')
```

```python
pd.merge(logins,reg,how='outer',on='name')
```

```python
pd.merge(reg,logins,how='left',on='name')
```

```python
pd.merge(reg,logins,how='right',on='name')
```
