# Multiple Conditions

1. &
2. |
3. ~ (NOT operator)

```python
df[~(df['total_bill'] > 35) & (df['sex']=='Female')]
```

```python
df[(df['total_bill'] > 35) | ~(df['sex']=='Female')]
```

```python
df[(df['total_bill'] > 35) | (df['sex']!='Female')]
```
