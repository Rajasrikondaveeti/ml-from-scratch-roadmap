```python
def unit_standardization(df):

  if df['Production Units'] == 'Nuts':
    return np.round(df['Production']/500,0)
  elif df['Production Units'] == 'Tonnes':
    return df['Production']
  else:
    return np.round(df['Production']/4.59296,0)
```

```python
df['New Production'] = df.apply(unit_standardization,axis=1)
df.sample(5)
```

```python
df[df['Production Units'] == 'Nuts']
```

```python
df = df.drop(columns=['Production','Production Units','Area Units'])
df.head()
```

```python
#Which crop is most common choice in India?

df['Crop'].value_counts().head(5)
```

```python
unique_states_list = df['State'].unique()
unique_states_list
```

## Get the statewise total crop production

```python
total_production = []

for st in unique_states_list:
  total_production.append(df.loc[df['State'] == st,'New Production'].sum())
```

```python
crop_production_df = pd.DataFrame(
    {
        'State' : unique_states_list,
        'Total Crop Production' : total_production
    }
)
```

```python
crop_production_df.sort_values('Total Crop Production', ascending=False).head(10)
```

```python
pd.set_option('display.float_format', '{:.8f}'.format)
```

## Which state is the highest producer of Maize between 1997 and 2022

```python
maize_df = df[df['Crop'] == 'Maize']

maize_states_list = maize_df['State'].unique()
maize_states_list
```

```python
total_maize_production = []

for st in maize_states_list:
  total_maize_production.append(maize_df.loc[maize_df['State'] == st,'New Production'].sum())
```

```python
maize_production_df = pd.DataFrame(
    {
        'State' : maize_states_list,
        'Total Crop Production' : total_maize_production
    }
)
```

```python
maize_production_df.sort_values('Total Crop Production', ascending=False).head(10)
```

# Best Year for Agriculture

```python
unique_year_list = df['Year'].unique()
```

```python
yearly_production = []

for yr in unique_year_list:
  yearly_production.append(df.loc[df['Year'] == yr, 'New Production'].sum())
```

```python
yearly_production_df = pd.DataFrame(
    {
        'year' : unique_year_list,
        'total_crop_production' : yearly_production
    }
)
```

```python
yearly_production_df.sort_values('total_crop_production',ascending=False).head()
```
