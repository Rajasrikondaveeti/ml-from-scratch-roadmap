## Methods

* functions unside the body of a class and used to perform operations with the attributes of our objects

```python
class Circle:
  pi = 22/7

  #class object attribute
  def __init__(self, radius = 1):
    self.radius = radius
    self.area = radius*radius*Circle.pi
    self.circumference = 2*Circle.pi*self.radius #Circle.pi = self.pi

  #Method
  def setRadius(self,new_radius):
    self.radius = new_radius
    self.area = new_radius*new_radius*Circle.pi

  def outputCircumference(self):
    return self.radius*self.pi*2
```

```python
c = Circle(radius=5)
```

```python
c.radius
```

```python
c.pi
```

```python
c.area
```

```python
c.setRadius(new_radius=9)
```

```python
c.radius
```

```python
c.area
```

```python
c.outputCircumference() #method
```

```python
c.circumference #attribute
```

```python
new_circle = Circle(radius=3)
```

```python
print(f"The radius of the circle is {new_circle.radius}")
print(f"The area of the circle is {new_circle.area}")
print(f"The circumference of the circle is {new_circle.circumference}")
```

```python
#Change the circumference

new_circle.setRadius(new_radius=8)
```

```python
print(f"The new radius of the circle is {new_circle.radius}")

print(f"The circumference of the circle, getting from the method, is {new_circle.outputCircumference()}")
```

#Create a Line class that accepts coordinates as a pair of tuples and return slope and distance of the line.

```python
class Line:
  def __init__(self, coordinate1, coordinate2):
    self.coordinate1 = coordinate1
    self.coordinate2 = coordinate2

  def distance(self):
    x1,y1 = self.coordinate1
    x2,y2 = self.coordinate2

    return ((x2-x1)**2 + (y2-y1)**2)**0.5

  def slope(self):
    x1,y1 = self.coordinate1
    x2,y2 = self.coordinate2
    return (y2-y1)/(x2-x1)
```

```python
coordinate1 = (9,8)
coordinate2 = (17,21)
```

```python
l1 = Line(coordinate1,coordinate2)
```

```python
l1.coordinate2
```

```python
l1.coordinate1
```

```python
l1.distance() #method
```

```python
l1.slope()
```

# create class Cylinder with methods/attrbutes : Volume, Surface Area

```python
class Cylinder:
  pi = 22/7

  def __init__(self,height=1,radius=1):
    self.height = height
    self.radius = radius

    self.volume = Cylinder.pi*self.height*(self.radius**2) #self.pi

  def surface_area(self):
    top = (self.radius**2)*Cylinder.pi
    return (2*top)+(2*Cylinder.pi*self.radius*self.height)
```

```python
cylinder1 = Cylinder(height=3,radius=4)
```

```python
cylinder1.volume
```

```python
cylinder1.surface_area()
```

#Inheritance

* Inheritance : class acquring properties and methods of another class

* Base Class (or Parent class) : The class being inherited from
* Derived Class (or Child class) : The class that inherits from the base class

```python
class Animal:
  def sound(self):
    return 'Some Sound'
```

```python
class Dog(Animal):
  def bark(self):
    return 'Woof !!'
```

```python
ani = Animal()
```

```python
ani.sound()
```

```python
ani.bark()
```

```python
dog1 = Dog()
```

```python
dog1.bark()
```

```python
dog1.sound()
```

```python
class Vehicle: #Base class / Parent class
  def __init__(self):
    print('Vehicle class created')

  def whoAmI(self):
    print('Vehicle')

  def eat(self):
    print('Eating Fuel')


class Cars(Vehicle): #Derived Class / Child Class

  def __init__(self):
    Vehicle.__init__(self)
    print('Car created')

  def whoAmI(self):
    print('Car')

  def sound(self):
    print('Porsche sound')
```

```python
v = Vehicle()
```

```python
v.sound()
```

```python
v.whoAmI()
```

```python
v.eat()
```

```python
c1 = Cars()
```

```python
c1.whoAmI()
```

```python
c1.eat()
```

# Super()


- allows to call a method from the parent class inside a child class

```python
class Human:
  def __init__(self,name,age=12):
    self.name = name
    self.age = age

  def hisName(self):
    return f'His age is {self.age}'


class Employee(Human):
  def __init__(self,name,employee_id,age):
    super().__init__(name,age) #call the __init__ of Human
    self.employee_id = employee_id
```

```python
sh = Human(name='Shreyas')
```

```python
sh.name
```

```python
ra = Employee(name='rahul',employee_id=101,age=21)
```

```python
ra.name
```

```python
ra.employee_id
```

```python
ra.age
```

```python
ra.hisName()
```
