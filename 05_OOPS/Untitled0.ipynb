```python
class Car:

  def __init__(self,name, company):
    self.name = name
    self.company = company
    self.available = True


  def __str__(self):
    return f"{self.name} by {self.company} ({'Available' if self.available else 'Rented'})"


class Boom:
  def __init__(self):
    self.cars = []


  def add_car(self,car):
    self.cars.append(car)

  def rent_car(self,name):
    for car in self.cars:
      if car.name == name and car.available:
        car.available = False
        return f"You rented '{car.name}'"
    return 'Car not available'

  def return_car(self,name):
    for car in self.cars:
      if car.name == name and not car.available:
        car.available = True
        return f"you returned '{car.name}'"
    return "Invalid Car"
```

```python
boom = Boom()
```

```python
c1 = Car('Slavia','Skoda')
```

```python
c2 = Car('Civic','Honda')
```

```python
boom.add_car(c1)
```

```python
boom.add_car(c2)
```

```python
for c in boom.cars:
    print(c)
```

```python
boom.rent_car('Slavia')
```
