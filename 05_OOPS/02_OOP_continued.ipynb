```python
class Dog:

  def __init__(self,name):
    self.name = name

  def speak(self):
    return self.name + ' says Woof Woof !!'


class Cat:
  def __init__(self,name):
    self.name = name

  def speak(self):
    return self.name + ' says Meow Meow !!'
```

```python
fredie = Dog(name='Fredie')
felix = Cat(name='Felix')
```

```python
print(fredie.speak())
print(felix.speak())
```

```python
#1. For loop method

for pet in [fredie, felix]:
  print(pet.speak())
```

```python
#2. Using functions

def pet_speaks(pet11):
  print(pet11.speak())
```

```python
pet_speaks(fredie)
```

```python
pet_speaks(felix)
```

```python
class Book:

  def __init__(self,title, author,pages):
    print('Congrats!! You have created a book')
    self.title = title
    self.author = author
    self.pages = pages

  def __str__(self):
    return f"Title : {self.title}\nPages : {self.pages}\nAuthor : {self.author}"

  def __len__(self):
    return self.pages

  def __del__(self):
    print('A book is destroyed')
```

Create a Bank Account class that enables deposits and withdrawal

```python
class BankAccount: #defines a new class
  def __init__(self, account_holder, balance=0):
    self.balance = balance
    self.account_holder = account_holder

  def deposit(self,amount):
    if amount <= 0:
      raise ValueError("deposit in Positives only")
    self.balance += amount
    return f'You have deposited {amount}. The updated balance is {self.balance}'

  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance -= amount
      return self.balance

    else:
      #return 'Insufficient Funds'
      raise ValueError("Insufficient Funds")

class SavingsAccount(BankAccount): #New class that inherits from BankAccount
  def __init__(self,account_holder ,balance = 0, interest_rate = 0.04):
    super().__init__(account_holder,balance) #call parent class (Bank Account) to initialize balance
    self.interest_rate = interest_rate

  def apply_interest_rate(self): #method to calculate and add interest to balance
    self.balance = self.balance + (self.balance*self.interest_rate)
    return self.balance

  def __str__(self):
    return f"Saving Account - Holder: {self.account_holder}\nBalance: {self.balance}\nInterest Rate : {self.interest_rate*100}%"
```

```python
sb = SavingsAccount(account_holder = 'Shreyas Shukla', balance=10000, interest_rate=0.06)
```

car class with attrbutes name, company, availablity

Boom car stroes a list of cars: Library should:

1. add cars
2. borrow a car by name if available
3. return a car

```python
class Car:

  def __init__(self,name, company):
    self.name = name
    self.company = company
    self.available = True


  def __str__(self):
    return f"{self.name} by {self.company} ({'Available' if self.available else 'Rented'})"


class Boom:
  def __init__(self):
    self.cars = []


  def add_car(self,car):
    self.cars.append(car)

  def rent_car(self,name):
    for car in self.cars:
      if car.name == name and car.available:
        car.available = False
        return f"You rented '{car.name}'"
    return 'Car not available'

  def return_car(self,name):
    for car in self.cars:
      if car.name == name and not car.available:
        car.available = True
        return f"you returned '{car.name}'"
    return "Invalid Car"
```

```python
boom = Boom()
```

```python
c1 = Car('Slavia','Skoda')
```

```python
c2 = Car('Civic','Honda')
```

```python
boom.add_car(c1)
```

```python
boom.add_car(c2)
```
