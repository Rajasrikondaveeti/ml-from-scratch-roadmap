```python
dataset = pd.read_csv('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/13_Linear_regression/Salary_Data.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

# Splitting the dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=50)
```

# Training the Simple Linear Rregression model on the Training Set

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() #Instantiate the LR class
```

```python
reg.fit(X_train,y_train)
```

# Predict the Test Results

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
y_test
```

Visualising the Test Set results

```python
plt.scatter(X_test,y_test,color='red')
plt.scatter(X_test,y_pred,color='green')
plt.plot(X_train,reg.predict(X_train), color='blue')
plt.title('Salary vs Experience')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```
