```python
df = pd.read_csv('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/13_Linear_regression/Advertising.csv')
df.head()
```

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

# polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures

p_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
p_features = p_converter.fit_transform(X)
```

```python
X.shape
```

```python
p_features.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(p_features,y,test_size=0.25,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc_X_train = sc.fit_transform(X_train)
sc_X_test = sc.transform(X_test)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
r_model = Ridge(alpha=10)
```

```python
r_model.fit(sc_X_train,y_train)
```

```python
r_test_pred = r_model.predict(sc_X_test)
```

#Ridge Regression Performance

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,r_test_pred)
MSE = mean_squared_error(y_test,r_test_pred)
RMSE = np.sqrt(MSE)
```

# Grid Search

* Search through a variety of combinations of hyperparameters with a grid search.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNet  #Base model
```

```python
param_grid = {'alpha': [0.1,1,5,10,50,100],
              'l1_ratio':[0,0.1,0.5,0.7,0.9,1],
              'max_iter' : [10000]}
```

```python
base_elastic_model = ElasticNet()
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid = param_grid,
                          cv=5,verbose=1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
y_pred = grid_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
