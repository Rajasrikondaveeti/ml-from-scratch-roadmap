```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
axes.set_title('x vs y')
plt.show()
```

# Add rows and columns

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
  ax.plot(x,y)
  x = x**2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[1].plot(a,b)

axes[0].plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12,8))

axes[1,1].plot(a,b)
axes[1,1].set_title('1 1 title')

axes[-1,-2].plot(x,y)
axes[-1,-2].set_xlabel('jkdasjka')

fig.suptitle('Figure level title',fontsize=50)
fig.supxlabel('Figure level x label here')

fig.subplots_adjust(wspace=0.9,hspace=1)

plt.show()
```
