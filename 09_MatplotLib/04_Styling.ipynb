```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
```

# legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='x**2')
ax.plot(x,x**3, label = 'x cubed')
ax.legend(loc=(1.1,0.02))
plt.show()
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig,axes = plt.subplots()

axes.plot(x,y, marker='+', ms= 20, linestyle='--', color='#634591',mew=7,mec='black', mfc = 'yellow',
          lw = 3)
axes.set_title('x vs y')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[1].plot(y,x, marker='o')

axes[0].plot(x,y,marker='s',color='black')
```
