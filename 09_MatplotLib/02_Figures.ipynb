```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a ** 4
b
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
x
```

```python
y
```

# Creating a Figure

```python
fig = plt.figure() #creating figure object (blank canvas)


#add set of axes to figure
axes = fig.add_axes([0,0,1,1]) #left,bottom, width, height

#plot on that set of axes
axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure(figsize=(8,3),dpi=100)

ax1 = fig.add_axes([0,0,1,1]) #Larger fig
ax2 = fig.add_axes([0.3,0.3,0.5,0.5]) #Smaller Fig
ax3 = fig.add_axes([1.1,1.1,0.3,0.3]) # Fig outside ax1

ax1.plot(a,b)
ax1.set_xlabel('A value here')
ax1.set_ylabel('B value here')
ax1.set_title('Big Figure')

ax2.plot(x,y)
ax2.set_xlabel('X value here')
ax2.set_ylabel('Y value here')
ax2.set_title('Small Figure')

ax3.plot(b,a)
ax3.set_xlabel('B value here')
ax3.set_ylabel('A value here')
ax3.set_title('Outside Figure')
ax3.set_xlim(0,5000)
ax3.set_ylim(2.5,5)


plt.savefig('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/09_MatplotLib/example_OOP.png',bbox_inches='tight')
plt.show()
```
