Assign grades based on avg and individual score

* 35 -->> F
* 90 -->> A
* 80 -->> B
* 70 -->> C
* between 35 and 70 -->> D

```python
marks = [90,87,67,49,29]

avg_marks = sum(marks)/len(marks)

if any(mark < 35 for mark in marks) == True:
  print('F')
elif avg_marks >= 90:
  print('A')
elif avg_marks >= 80:
  print('B')
elif avg_marks >= 70:
  print('C')
else:
  print('D')
```

Print "Even Prime" for 2,
"Even" for even numbers,
"Odd Prime" for odd primes
"odd" otherwise for numbers 1 to 100

```python
for num in range(1,101):
  if num == 2:
    print(f'{num} is Even Prime')
  elif num%2 ==0:
    print(f"{num} is even")
  else:
    prime = True
    for n in range(2,num):
      if num%n == 0:
        prime = False
        break
    if prime and num != 1:
      print(f'{num} is Odd Prime')
    else:
      print(f'{num} is Odd')
```

## Sum numbers from 1 to N excluding multiples of 3 or 7

```python
num = int(input('Enter a number: '))
total = 0
for n in range(1,num+1):
  if n%3 ==0 or n%7 == 0:
    continue
  total = total+n

print(total)
```

Count vowels, consonants and digits

```python
s = 'Hello 1234MASAI**^^*'

vowels = {'a','e','i','o','u'}

v = 0
c = 0
d = 0
r = 0
for char in s:
  if char.isalpha():
    if char.lower() in vowels:
      v = v+1
    else:
      c = c+1
  elif char.isdigit():
    d += 1
  else:
    r = r+1

print(f'Vowels: {v}\nConsonants: {c}\nDigits: {d}\nRemaining characters: {r}')



```
