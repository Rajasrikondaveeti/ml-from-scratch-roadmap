```python
#Check is a password has atleast 8 characters, 1 uppercase, 1 special characters and 1 digit

password = input('Enter your password here: ')
errors = []

if len(password) < 8:
  errors.append("Password Too short")
if not any(c.isupper() for c in password):
  errors.append("Missing Uppercase")
if not any(c.isdigit() for c in password):
  errors.append("Digit not found")
if not any(c in '!@#$%^&*' for c in password):
  errors.append('Missing special characters')


print('Valid' if not errors else 'Invalid: '+", ".join(errors))
```

```python
password = "test1234"

has_upper = False
has_digit = False

if len(password) >= 8:
    for char in password:
        if char.isupper():
            has_upper = True
        if char.isdigit():
            has_digit = True

    if has_upper and has_digit:
        print("Valid.")
    else:
        print("Password must contain at least one uppercase and one digit.")
else:
    print("Password must be 8 characters long.")
```

#### Write a program that prints integers from 1 to 100.

* But for multiples of 3 : 'Shreyas'
* For multiples of 7 : 'MASAI'
* for multiples of both 3 and 7 : 'IITG'

```python
for num in range(1,101):
  if num%3 == 0 and num%7 == 0:
    print('IITG')
  elif num%3 == 0:
    print('Shreyas')
  elif num%7 == 0:
    print('MASAI')
  else:
    print(num)
```
