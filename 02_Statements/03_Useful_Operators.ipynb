## Enumerate

```python
idx_count = 0

for l1 in 'shreyas':
  print(f'At index {idx_count} the letter is {l1}')
  idx_count = idx_count + 1
```

```python
for i,l in enumerate('shreyas'):
  print(f'At index {i} the letter is {l}')
```

```python
list(enumerate('shreyas'))
```

## Zip

```python
l1 = [21,22,23,24,25]
l2 = ['a','b','c','d','e']
```

```python
list(zip(l1,l2))
```

```python
for i1, i2 in zip(l1,l2):
  print(f'For this tuple, first item was {i1}, and second item was {i2}')
```

```python
#in

'a' in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
34.9 in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
'shreyas' in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
['shreyas','hello'] in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
#not in

'a' not in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
['shreyas','hello'] not in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```
