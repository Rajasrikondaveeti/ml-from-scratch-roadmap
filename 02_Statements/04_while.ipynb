### break : break out of the current closest enclosing loop
### continue : goes to the top of closest enclosing loop
### pass : placeholder. Does nothing

```python
x = 0

while x < 10:
  print('x is currently : ', x)
  print('adding 1 to x')
  x = x+ 2
  if x == 3:
    print('breaking because x == 3')
    break
  else:
    print('continuing')
    continue
```

```python
# Check a number is a palindrome

num = int(input('Enter your number here:'))

original_no = num

reverse = 0

while num > 0:
  d = num % 10
  reverse = reverse * 10 + d
  num = num // 10

print("Number is palindrome" if original_no == reverse else " It is not a palindrome")
```

```python
#n --> while loop to calculate sum of first n natural numbers

n = 9

i = 1
total = 0

while i <= n:
  total += i
  i += 1

print(f'Sum: {total}')
```
