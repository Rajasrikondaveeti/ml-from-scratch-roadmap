```python
#while (repeats while a condition holds)


n = 9

while n > 0:
  if n == 3:
    n -= 1
    continue #skips 3
  print(n**2)
  n -= 1
while True:
  if n == 0:
    break #exits when n = 0
  n -= 1
```

```python
#Functions

def add11(a,b, *args, **kwargs):
  total = a+b+ sum(args)
  if 'scale' in kwargs:
    total *= kwargs['scale']
  return total
```

```python
add11(1,2,3,2,1,2,3,1,32,4,2,3,2,1,34,2,4,2,4,2,2,scale=2)
```

## OOP

* Classes define templates
* Objects are instances
* __init__ initializes instance variables
* Methods define behaviour
* Attributes define characteristics

```python
class Car:
  wheels = 4 #class variable
  def __init__(self,brand):
    self.brand = brand #instance variable

  def new_car(self,updated_brand):
    self.brand= updated_brand

  def drive(self):
    return f'{self.brand} is driving !!'
```

```python
c1 = Car(brand='Toyota')
```

```python
c1.brand
```

```python
c1.wheels
```

```python
c1.drive()
```

```python
c2 = Car(brand='Suzuki')
```

```python
c2.brand
```

```python
c2.wheels
```

```python
c2.new_car(updated_brand='Mahindra')
```

```python
c2.brand
```

```python
c2.wheels
```

```python
c2.drive()
```

# inheritance

```python
class Vehicle: #base class
  def move(self):
    return 'Moving'

class Bike(Vehicle): #derived class (inheriting all the properties from the base class)
  def move(self):
    return f'{super().move()} on two wheels'
```

```python
b1 = Bike()
```

```python
b1.move()
```
