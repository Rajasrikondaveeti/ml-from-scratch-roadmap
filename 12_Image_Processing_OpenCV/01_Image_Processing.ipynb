```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate an image

#create a 1-D array of all the pixel values from 0 to 255
im = np.arange(0,256)


#im.shape

#Convert it to 2-D
im = im.reshape(1,im.shape[0])

#im.shape

#Repeat the rows 100 times (vertically)
im = np.repeat(im,100,axis=0)

im
```

```python

plt.imshow(im, cmap='gray')
```
