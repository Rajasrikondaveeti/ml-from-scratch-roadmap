## Factorial

```python
def factorial(n):
  res = 1
  for i in range(1,n+1):
    res *= i
  return res
```

```python
x = factorial(5)
```

'freezing' if temp is 0 or below

'cold' is between 1 and 15

'warm' if it is between 16-25

'hot' otherwise


temp in fahrenheit


C = (F-32)*5/9

```python
def check_temp(Fah):
  C = (5/9)*(Fah-32)

  if C <= 0:
    return f'The temp {Fah} degree farenheit or {C} degree C is Freezing'
  elif C <= 15:
    return f'Cold'
  elif C <= 25:
    return f'Warm'
  else:
    return f'Hot'

```

```python
check_temp(30)
```

(n1,n2,operator)

operator = +,-,*,/

```python
def calculator(n1,n2,op):
  if op == '+':
    return n1+n2
  elif op == '-':
    return n1-n2
  elif op == '*':
    return n1*n2
  elif op == '/':
    return n1/n2
  else:
    return 'Invalid Operator given'
```

```python
calculator(34,2,'%')
```

```python
def find_anagram(word,list_words):

  return [w for w in list_words if sorted(w) == sorted(word)]
```

```python
find_anagram('listen',['silent','google','inlets','enlist','banana'])
```

function that rotates a list by k elements to the right

```python
def rotates(lst , k):
  k %= len(lst)
  return lst[-k:]+lst[:-k]


#rotates([1,2,3,4,5],2)-->> [4,5,1,2,3]
```

```python
rotates([1,2,3,4,5],7)
```

transpose a matrix

```python
[[1,2,3],[4,5,6],[7,8,9]] -->> [[1,4,7],[2,5,8],[3,6,9]]
```

```python
def transpose(matrix):
  return [[row[i] for row in matrix] for i in range(len(matrix[0]))]
```

```python
mat = [[1,2,3,4],[4,5,6,7],[7,8,9,10]]
transpose(mat)
```
