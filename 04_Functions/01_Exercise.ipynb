```python
# 1. Find the maximum of element-wise sums using zip, map , lambda

a = [1,2,3,4,5,6,7]
b = [21,32,43,44,55,66,87]
```

```python
max(list(map( lambda x : x[0] + x[1] ,zip(a,b))))
```

```python
#2 return list of squared odd numbers from 1 to 50 using list comprehension and lambda with filter

[x**2 for x in filter(lambda x : x%2 != 0,range(1,51))]
```

```python
#3 Write a function that categories a number as 'small', 'medium', 'large' using if-elif-else and lambda

def categorize(nums):
  return list(map(lambda x : 'small' if x < 10 else 'medium' if x<100 else 'large',nums))
```

```python
categorize([12,13,54,28,103])
```

```python
#5 Write a while loop that keeps summing random number from a list until total > 50

l1 = [5,8,12,3,10,9,23,34,2,4,13,98]

i = 0

total = 0

while total < 51 and i < len(l1):
  total += l1[i]
  i += 1

print(f"Total: {total}")
```

```python
#7 Create a function that counmts how many times each vowel appears

def count_vowels(s):
  return {v : s.lower().count(v) for v in 'aeiou'}
```

```python
count_vowels('Daksh Gurukul IIT Guwahati')
```

```python
#8 Create a function that finds the word with most vowels


def word_with_most_vowels(word):
  vowels = 'aeiou'
  max_count = 0
  resultant_word = ''

  for w in word:

    count = sum(1 for c in w.lower() if c in vowels)
    if count > max_count:
      max_count = count
      resultant_word = w

  if max_count == 0:
    return 'No vowels found in any of the given word'

  return resultant_word
```

```python
word_with_most_vowels(['strength','Stamina','IT','Gurukul','Shreyas'])
```

```python
word_with_most_vowels(['sky','gym','Dry'])
```
