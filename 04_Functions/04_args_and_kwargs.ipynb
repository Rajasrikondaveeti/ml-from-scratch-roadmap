## **kwargs

```python
def any_func(**kwargs):
  if 'fruit' in kwargs:
    print(f"My fav fruit is {kwargs['fruit']}")
  else:
    print('I dont like fruit')
```

```python
any_func(fruit='pineapple')
```

```python
any_func(fruit='pineapple', veggie  ='carrot')
```

```python
def demo_func(*args, **kwargs):
  if 'fruit' and 'juice' in kwargs:
    print(f"I like { ' and '.join(args)}, my fav fruit is {kwargs['fruit']}")
    print(f"May I have some {kwargs['juice']} juice?")
  else:
    pass
```

```python
demo_func('eggs','spam','biscuits','bread', fruit = 'mangoes', juice = 'orange')
```

```python
#*args (it is a tuple)

def add_numbers(*args):
  total = sum(args)

  print('Arguments: ',args)
  print('Total: ', total)
```

```python
add_numbers(5,10,12)
```

```python
#**kwargs (it is a dictionary):

def say_namaste(**kwargs):
  for k,v in kwargs.items():
    print(f"{k} : {v}")
```

```python
say_namaste(name='Pulkit',age=28, city='Rome')
```

```python
for k , v in {'name':'Pulkit'}.items():
  print(f"{k} and {v}")
```

```python
def booking_summary(date, *passengers, **details):
  print(f"Booking Date: {date}")
  print("Passenger Lists:")
  for p in passengers:
    print("-",p)
  print("Additional Details:")
  for key,value in details.items():
    print(f"{key} : {value}")
```

```python
booking_summary("2025-06-03",
                "Kuldeep","Ritul","Vipan","Viraj",
                flight_no = "AI-1034",
                class1 = 'Business',
                meal = "Veg")
```

```python
def describe_animals(name, species):
  print(f'{name} is a {species}')
```

```python
args = ('Max','Dog')
describe_animals(*args)
```

```python
kwargs = {'name':'Bunny','species':'cat'}
describe_animals(**kwargs)
```
