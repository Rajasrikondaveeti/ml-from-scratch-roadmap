# Rules

1. Name assignments will create or change local names by default
2. Name references search four scopes at most: local, enclosing functions, global, Built-in (LEGB)

* Local : Names assigned in any way within a function and not declared in that function
* Enclosing function locals - Names in the local scope of any and all enclosing function
* Global(Module) - Names aasigned at the top level of a module file or decalred global in a def within a file
* Buil-In - Names pre-assigned in the built-in names modele (range, enumerate etc.)

```python
#Local

def func():
  x = 10 #Local Scope
  return x
```

```python
func()
```

```python
#Enclosing

def outer():
  x = 20 #Enclosing Scope

  def inner():
    print(x) #Python looks here if not found in local

  inner()

outer()
```

```python
inner()
```

## Local Variables

```python
x = 50

def func(x):
  print('x is',x)
  x = 2 #local
  print('Changed local x to',x)

func(x)
print('Global x is still', x)
```

```python
#Global statement

x = 50

def func():
  global x
  print('This funciton is now using global x')
  print('Because of global x is ',x)
  x = 2
  print(f'Ran func() , changes global x to {x}')
```

```python
print('before calling func(), x is:',x)
```

```python
func()
```
