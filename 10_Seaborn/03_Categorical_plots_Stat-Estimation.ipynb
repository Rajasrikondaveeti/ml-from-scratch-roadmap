```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/10_Seaborn/dm_office_sales.csv')
df.head()
```

```python
# Countplot
plt.figure(figsize=(12,4))
sns.countplot(x='division',data=df,palette='Set1',hue='training level') #visual alternative of value counts
```

```python
#barplot
plt.figure(figsize=(12,4))
sns.barplot(x='division',y='salary',data=df,palette='Set1',estimator=np.var,hue='level of education')
plt.title('This is barplot')
plt.xlabel('Division in the Firm')
plt.ylabel('Salary of employees')
plt.xticks(rotation=90)
plt.legend(loc=(1.01,0.01))
plt.show()
```
