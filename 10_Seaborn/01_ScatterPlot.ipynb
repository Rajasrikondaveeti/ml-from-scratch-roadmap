# Scatterplot

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='division',palette='husl')
plt.show()
```

color palette: https://seaborn.pydata.org/tutorial/color_palettes.html

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='rocket_r')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='division',palette='husl',size='training level',
                style='level of education')
plt.legend(loc=(1.01,0.05))
plt.show()
```
