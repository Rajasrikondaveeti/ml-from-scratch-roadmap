```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

```python

df = pd.read_csv('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/10_Seaborn/StudentsPerformance.csv')
df.head()
```

# Joint Plot

```python
plt.figure(figsize=(5,5))
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
plt.figure(figsize=(5,5))
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
plt.figure(figsize=(5,5))
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```
