```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

```python

df = pd.read_csv('/content/drive/MyDrive/Daksh_Gururkul_IITG_May_2025/10_Seaborn/StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

# Box Plot

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df, x='parental level of education',y='math score',hue='gender')
```

```python
#Orientation

plt.figure(figsize=(12,5))
sns.boxplot(data=df, y='parental level of education',x='math score',hue='gender',orient='h')
```

# Violin Plot

```python
plt.figure(figsize=(12,3))
sns.violinplot(data=df, x='parental level of education',y='math score',hue='gender',inner='box',split=True)
plt.legend(loc=(1.01,.01))
```

```python
#help(sns.violinplot)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, y='parental level of education',x='math score',hue='gender',inner='box',split=True)
plt.legend(loc=(1.01,.01))
```

# Swarmplots

```python
sns.swarmplot(x='math score',data=df,size=3)
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df,size=3, hue='gender',dodge=True)
```
